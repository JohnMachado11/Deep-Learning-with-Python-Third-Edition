## Data Representations for Neural Networks

### Scalars (rank-0 tensors)

```python
import numpy as np

x = np.array(12)
x
```

```python
x.ndim
```

### Vectors (rank-1 tensors)

```python
x = np.array([12, 3, 6, 14, 7])
x
```

```python
x.ndim
```

### Matrices (rank-2 tensors)

```python
x = np.array([[5, 78, 2, 34, 0],
              [6, 79, 3, 35, 1],
              [7, 80, 4, 36, 2]])
x
```

```python
x.ndim
```

```python
x.sum(axis=0) # sum across rows 
# 5 + 6 + 7 = 18
```

```python
x.sum(axis=1) # sum across columns
# 5 + 78 + 2 + 34 + 0 = 119
```

#### Rank-3 tensors and higher-rank tensors

```python
x = np.array([[[5, 78, 2, 34, 0],
               [6, 79, 3, 35, 1],
               [7, 80, 4, 36, 2]],
              [[5, 78, 2, 34, 0],
               [6, 79, 3, 35, 1],
               [7, 80, 4, 36, 2]],
              [[5, 78, 2, 34, 0],
               [6, 79, 3, 35, 1],
               [7, 80, 4, 36, 2]]])
```

```python
x.ndim
```

```python
x.shape
```

#### Manipulating tensors in NumPy

```python
my_slice = train_images[10:100]
my_slice.shape
```

```python
# Equivalent to the previous example
my_slice = train_images[10:100, :, :]
my_slice.shape 
```

```python
# Also equivalent to the previous example
my_slice = train_images[10:100, 0:28, 0:28]
my_slice.shape
```

```python
#o select 14  14 pixels in the bottom-right corner of all images, you would do this:
my_slice = train_images[:, 14:, 14:]
```

```python
# my_slice[0]
```

```python
# To crop the images to patches of 14  14 pixels centered in the middle, do this:
my_slice = train_images[:, 7:-7, 7:-7]
```

```python
# my_slice[0]
```
